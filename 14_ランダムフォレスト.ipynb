# ランダムフォレストによるアヤメの分類

```python
#! pip install pydotplus
import seaborn as sns
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestRegressor
from sklearn.externals.six import StringIO
import pydotplus
import matplotlib.pyplot as plt
```

```python
# Iris データの読込と表示
# 変数:4、サンプル数:150、種類:3
iris = load_iris()
print(iris.DESCR)
```

```python
# Irisデータを表示
iris_view = sns.load_dataset('iris')
iris_view
```

```python
# 入力変数と説明変数の切り分け
X = iris.data  
Y = iris.target
```

```python
# ランダムフォレストによる分類の実行 (デフォルトでは決定木を9つ生成;n_estimators=10)
clf = RandomForestRegressor()
clf.fit(X, Y)  
```

```python
# それぞれの決定木の可視化
forest = clf.fit(X, Y)  
from pylab import rcParams
rcParams['figure.figsize'] = 20,10
for i,val in enumerate(forest.estimators_):
    tree.export_graphviz(forest.estimators_[i], 
                         out_file='tree_%d.dot'%i, 
                         feature_names=iris.feature_names, 
                         class_names=iris.target_names, 
                         filled=True, 
                         rounded=True,
                         special_characters=True)
    graph = pydotplus.graph_from_dot_file("tree_"+ str(i)+".dot") 
    graph.write_png("tree_"+ str(i)+".png") 
    img = plt.imread("tree_"+ str(i)+".png")
    plt.title("tree_"+ str(i)+".png")
    plt.imshow(img)
    plt.show()   
    
# 出力ファイルの削除
from pathlib import Path
for p in Path("").glob("tree_*.*"):
    p.unlink()
```

```python
# 各要素の重要度(影響力の高いもの)を表示 [sepal_length, sepal_width, petal_length, petal_width]
clf.feature_importances_ 
```

```python
# 予測精度の確認
clf.score(X,Y)
```

```python
# 元データの150件目(X[149])で分類結果を表示(0=Setosa,1=Versicolor, 2=Virginica)
print (clf.predict([X[149]]))
```

```python
# 元データの150件目(Y[148])の正解を表示(0=Setosa,1=Versicolor, 2=Virginica)
print ([Y[149]])
```
